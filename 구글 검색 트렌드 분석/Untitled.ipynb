# 구글 트렌드 시계열 그래프

```python
from pytrends.request import TrendReq
from matplotlib import pyplot as plt 	# 그래프를 그리기 위한 라이브러리
%matplotlib inline 	
```

```python
# 검색 키워드, 기간 설정
keyword = "COVID-19"
period = "today 3-m" # 검색 기간: 최근 3개월

# Google Trend 접속
trend_obj = TrendReq()
trend_obj.build_payload(kw_list = [keyword], timeframe = period)
```

```python
# 시간에 따른 Trend 변화
timeline_df = trend_obj.interest_over_time()
print(timeline_df.head())
```

```python
# 그래프 출력
plt.figure(figsize = (10, 4))
timeline_df["COVID-19"].plot()
plt.title("Google Trend Over Time", size = 15)
plt.legend(labels = ["COVID-19"], loc = "upper right") # 그래프 우측 상단 라벨명 표시
```
